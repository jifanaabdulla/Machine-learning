### Logistic regression using titanic dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('titanic.csv')
data
```

```python
#check the null values
data.isnull().sum()
```

```python
#drop the null value column
data.drop('Cabin',axis=1,inplace=True)#inplace=True'  make changes permanent 
data
```

```python
#to fill next null column 
data.Age.fillna(data.Age.median(),inplace=True)
```

```python
#drop the null values
# data = data.dropna()
# data
```

```python
data
```

```python
#check the null values
data.isnull().sum()
```

```python
#remove that null row fare
data.dropna(inplace=True)
data
```

```python
#check the null values
data.isnull().sum()
```

```python
#check shape
data.shape
```

```python
data.groupby ('Survived').size()
```

```python
cleaned_data = data.drop (columns = ['Name','Ticket'], axis = 1)
```

```python
cleaned_data
```

#### To apply logistic regression all the columns should have categorial or numerical values. 

```python
df=pd.get_dummies (cleaned_data ['Sex'], dtype=int)
```

```python
df1 = pd.get_dummies (cleaned_data ['Pclass'],drop_first = True,dtype=int)#drop_first=True  it helps in reducing the extra column created during dummy variable creation.
```

```python
df1
```

```python
df2 =pd.get_dummies (cleaned_data ['Embarked'],dtype=int)
df2
```

```python
modified_data= pd.concat([cleaned_data, df, df1,df2], axis = 1)
```

```python
modified_data
```

```python
modified_data.info()
```

```python
y= modified_data ['Survived']
x= modified_data.drop (['Sex','Survived','PassengerId','Embarked'], axis = 1 )
```

```python
x
```

```python
y
```

```python
x.columns = x .columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test =  train_test_split (x,y,test_size=0.1)
```

```python
x_train
```

```python
x_test
```

```python
model =LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
#accuracy of a model
model.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(classification_report(y_test,y_predict))
```

```python
cm =confusion_matrix(y_test,y_predict)
```
